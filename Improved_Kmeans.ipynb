```python
 # Calculate Euclidean distance.
def get_distance(a,b,axis=1):
    return np.linalg.norm(a-b)
```

```python
# calculate AvgDistance
def AvgDist(a):
    sum=0;
    for i in range(len(a)):
        for j in range(len(a)):
            if(j > i):
                #print(i,j)
                sum=sum+get_distance(a[i],a[j])
        #print(" ")
    #print("total sum is:",sum)
    AvgDist=(2/(len(a)*(len(a)-1)))*sum
    
    return AvgDist
```

```python
k=AvgDist(X)
k
```

```python
epsilon=0.3*k
epsilon
```

```python
# calculating density parameter
def DensityParameter(a):
    S=[]
    
    for i in range(len(a)):
        total=0
        for j in range(len(a)):
            if(i != j):
                sum=epsilon - get_distance(a[i],a[j])
                if (sum >=0):
                    sum=1
                else:
                    sum=0
                total=total+sum
        S.append(total)
    #index_max = np.argmax(S)
    #print (S)
    return S#a[index_max]       
    
```

```python
# To store the value of centroids when it updates
C_old = np.zeros(C.shape)
print(C_old)
```

```python
# Cluster Lables(0, 1, 2)
clusters = np.zeros(len(X))
print(clusters)
```

```python
# Error func. - Distance between new centroids and old centroids
error = get_distance(C, C_old, None)
print (error)
```

```python
from copy import deepcopy
```

```python
k=2
```

```python
# Loop will run till the error becomes zero
while error != 0:
    # Assigning each value to its closest cluster
    for i in range(len(X)):
        distances = get_distance(X[i], C)
        cluster = np.argmin(distances)
        clusters[i] = cluster
    # Storing the old centroid values
    C_old = deepcopy(C)
    # Finding the new centroids by taking the average value
    for i in range(k):
        points = [X[j] for j in range(len(X)) if clusters[j] == i]
        C[i] = np.mean(points, axis=0)
    error = get_distance(C, C_old, None)

colors = ['r', 'g', 'b', 'y', 'c', 'm']
fig, ax = plt.subplots()
for i in range(k):
        points = np.array([X[j] for j in range(len(X)) if clusters[j] == i])
        ax.scatter(points[:, 0], points[:, 1], s=7, c=colors[i])
ax.scatter(C[:, 0], C[:, 1], marker='*', s=200, c='#050505')
```
